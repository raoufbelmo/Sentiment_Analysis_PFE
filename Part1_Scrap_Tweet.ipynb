### Scrap tweet with "twint" and filtered the language


```python
# link of tools twint 'https://github.com/twintproject/twint'
import os
import twint
import nest_asyncio
import pandas as pd
nest_asyncio.apply()

#fonction scrap

def scrap_tweet(keyword,namefile, start=None, end=None, batch_size=250, tempfile='temp.csv'):
    '''
    keyword : domain name search
    namefile : file name
    start : start date
    end : end date
    batch_size : number of tweet
    
    '''
    
    c = twint.Config()
    c.Search = keyword
    c.Limit = batch_size
    c.Since = start
    c.Until = end
    c.Store_csv = True
    c.Hide_output = True
    c.Output = tempfile

    twint.run.Search(c)

    try:
        data = pd.read_csv(tempfile)
        df = data[['date','time','username','tweet','language']]
        df.to_csv(namefile)
        
        if os.path.exists(tempfile):
            os.remove(tempfile)
        return df
    except FileNotFoundError:
        print('No search result is returned.')
        
```

```python
data=scrap_tweet(keyword='covid19',namefile='covid2021.csv')
data
```

```python
#filtered the language
data=data[lambda df:data['language']=='en' ]
data
```
